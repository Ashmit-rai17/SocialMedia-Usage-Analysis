Bivariate / Correlation Analysis

Overall, these trends suggest that stress, digital habits, and sleep quality are closely intertwined, and together they significantly influence ones overall sense of happiness and well-being.

```python
df[df["Age"].notnull()]
```

```python
def GroupAge(x):
    if x < 18:
        return "Teenager"
    elif 18 <= x < 30:
        return "Adult"
    elif 30 <= x < 45:
        return "Middle-Aged"
    else:
        return "Old"
df["GroupAge"] = df["Age"].apply(GroupAge)
```

```python
df["GroupAge"].value_counts()
```

```python
pd.crosstab(df['GroupAge'], df['Social_Media_Platform'], normalize='index').plot(kind='bar')
```

This is a more group-based percentage distribution according to each age group as we implemented the crosstab function which is essentially groupby funtion mixed with unstack and normalization funtion in .apply()
These are the following observations , we observed in the dataset:


Adults and Middle-Aged groups appear to have relatively balanced platform preferences (diversified social media usage). Adults having a bias towards platforms like X.

Teenagers might show stronger relative proportions for TikTok and Instagram, consistent with real-world trends.

Older groups likely lean toward Facebook and LinkedIn, showing more conservative platform choices

```python
df.groupby('GroupAge')[['Exercise_Frequency(week)','Stress_Level(1-10)']].corr()
```

Summary of Age Group Analysis

Across different age demographics, clear behavioral distinctions emerge. Teenagers show the lowest overall stress levels and a moderate screen-time pattern, with a negative correlation between exercise and stress  indicating that physical activity strongly contributes to emotional balance in this group. Adults display moderate stress and exercise levels, reflecting a relatively balanced but somewhat inconsistent lifestyle. Middle-aged individuals experience comparable median stress to adults but exhibit a wider spread, suggesting greater lifestyle variability  possibly due to work or family responsibilities. Interestingly, older participants demonstrate the highest median exercise frequency yet also the highest stress levels and variability. This counterintuitive finding may suggest that, for older individuals, stress could stem less from inactivity and more from age-related health or social factors, highlighting that physical activity alone doesnt necessarily equate to lower stress in this demographic.
